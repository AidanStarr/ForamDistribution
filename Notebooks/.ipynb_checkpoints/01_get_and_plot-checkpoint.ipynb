# 01: Load the dataset and visualize it 

## Intro
- Using the ForCenS database of sediment surface planktonic foraminifera by Siccha and Kucera 
>  Siccha, M; Kucera, M (2017): ForCenS, a curated database of planktonic foraminifera census counts in marine surface sediment samples. Scientific Data, 4, 170109, https://doi.org/10.1038/sdata.2017.109
- pull this dataset from Pangaea.de 
- start plotting it

----


```python
# Preamble
import numpy as np
import pandas as pd
import xarray as xr
import cartopy as ctpy
import matplotlib.pyplot as plt
```

## Pull the dataset
The data is available at pangaea.de as full or reduced (following the steps in the paper). We'll download the reduced...

> https://doi.org/10.1594/PANGAEA.873570

```python
df = pd.read_csv('https://store.pangaea.de/Publications//Siccha-Kucera_2017/ForCenS_reduced.txt',sep='\t',header=0) # download data from Pangaea.de repository
df.drop(index=0,inplace=True) # drop the units row - can retrieve later if needed
df.head() # preview data
```

```python
# change the datatypes of numeric columns to floating point
for c in df.columns[4:]:
    df[c] = df[c].apply(pd.to_numeric,errors='ignore')
    

df.dtypes # check that worked
```

```python
crs = ctpy.crs.EckertI()
# Plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection=crs);
ax.add_feature(ctpy.feature.COASTLINE);
dpth = ax.scatter(df.Longitude,df.Latitude,c=df['Water_depth'],transform=ctpy.crs.PlateCarree(),s=10,cmap='viridis_r');
fig.colorbar(dpth,label='Water Depth (m)',pad=0.01,fraction=0.035,orientation='horizontal');
ax.set_title('ForCenS Sample Depths',fontweight='bold');
```

```python
crs = ctpy.crs.Robinson()
# Plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection=crs)
ax.add_feature(ctpy.feature.COASTLINE)
cc = 'Globorotalia_truncatulinoides_sinistral_coiling'
sc = ax.scatter(df.Longitude,df.Latitude,transform=ctpy.crs.PlateCarree(),s=df[cc]*400,edgecolor='k')
ax.set_global()
ax.set_title('G. truncatulinoides (sin) Relative Abudnance',fontweight='bold');
```

```python
# save dataframe for later use
df.to_csv('../data/ForCenS_cleaned.csv',index=False)

# save G truncatulinoides for DIVA gridding
df_sorted = df.sort_values('Latitude')
df_diva = pd.DataFrame(columns=['lon','lat','G_trunc_s'])
df_diva['lat'] = df_sorted['Latitude']
df_diva['lon'] = df_sorted['Longitude']
df_diva['G_trunc_s'] = df_sorted['Globorotalia_truncatulinoides_sinistral_coiling']
df_diva.dropna().to_csv('../data/Gtruncs_for_diva.txt',sep='\t',index=False,header=False)
```

## Compare to hydrographic data
- use the GLODAPv2 data product


```python
# Temperature from GLODAP mapped climatology (provided in /data)
T = xr.open_dataset('../data/GLODAPv2.2016b.temperature.nc')
T
```

```python
crs = ctpy.crs.Robinson()
# Plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1, 1, 1, projection=crs)
ax.add_feature(ctpy.feature.COASTLINE)

tplot = T['temperature'].sel(depth_surface=0).plot(ax=ax,transform=ctpy.crs.PlateCarree(),add_colorbar=False,vmin=-10,vmax=35,cmap='RdBu_r')
fig.colorbar(tplot,fraction=0.035,orientation='horizontal',pad=0.05,label='Temperature ($^oC$)')

cc = 'Globorotalia_truncatulinoides_sinistral_coiling'
sc = ax.scatter(df.Longitude,df.Latitude,transform=ctpy.crs.PlateCarree(),s=df[cc]*500,edgecolor='k',c=df[cc],cmap='Greens')
ax.set_global()
ax.set_title('G. truncatulinoides (sin) Relative Abudnance',fontweight='bold');
```
